```python
import math
from sympy import *
from sympy.plotting import plot
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = Symbol('x')
fx = -12 * x ** 4 * sin(cos(x)) - 18 * x ** 3 + 5 * x **2 + 10 * x - 30
print(fx)
```

```python
pr = diff(fx)
print(pr)
```

```python
solveset(fx)
```

```python
trigsimp(pr)
solve(pr > 0) # интервал на котором функция возрастает
```

```python
solve(pr < 0) # интервал на котором функция убывает
```

```python
p = plot(fx)
```

```python
apex = solve(diff(fx))  
print(*apex) 
#вершина
```

```python
solve(fx > 0) # функция больше 0
```

```python
solve(fx < 0) # функция меньше 0
```
