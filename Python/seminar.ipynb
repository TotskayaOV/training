```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func(a,b,c,x):
    return a * x ** 2 + b * x + c
```

```python
def disc(a, b, c):
    return b**2 - 4* a * c
```

```python
x = np.arange(-10,10, 0.01)
a, b, c = 5, 10, -30
```

```python
def find_sqrt(a, b, c):
    d = disc(a, b, c)
    if d < 0:
        return 'Корней нет'
    elif d == 0:
        x = -b / (2 * a)
        return (round(x, 2))
    else:
        x1 = (-b + d **0.5)/ (2 * a)
        x2 = (-b - d **0.5)/ (2 * a)
        return (round(x1, 2), round(x2, 2))
```

```python
scrts = find_sqrt(a, b, c)
```

```python
min_y = min(func(a, b, c, x))
min_x = find_sqrt(a, b, (c-min_y))
```

```python

grow_up =np.arange(min_x, 10, 0.01)
grow_down = np.arange(-10, min_x, 0.01)
```

```python
pos_down =np.arange(-10, scrts[1], 0.01)
pos_up = np.arange(scrts[0], 10, 0.01)
neg_down = np.arange(scrts[1], min_x, 0.01)
neg_up = np.arange(min_x, scrts[0], 0.01)
```

```python
# plt.plot(x, func(a,b,c,x), 'y-')
# корни
plt.plot(scrts[0], 0, 'ro', label = f'1 корень: {scrts[0]}')
plt.plot(scrts[1], 0, 'go', label = f'2 корень: {scrts[1]}')
# вершина
plt.plot(min_x, min_y, 'bx', label = f'Вершина {min_x, min_y}')
# возрастание графика
# plt.plot(grow_up, func(a,b,c,grow_up), 'b-')
# убывание графика
# plt.plot(grow_down, func(a,b,c,grow_down), 'r-')
plt.rcParams['lines.linestyle'] = '-'
plt.plot(pos_down, func(a,b,c,pos_down), 'b', label = 'Убывает больше 0')
plt.rcParams['lines.linestyle'] = '-'
plt.plot(pos_up, func(a,b,c,pos_up), 'y', label = 'Возрастает больше 0')
plt.rcParams['lines.linestyle'] = '-.'
plt.plot(neg_up, func(a,b,c,neg_up), 'y', label = 'Возрастает меньше 0')
plt.rcParams['lines.linestyle'] = '-.'
plt.plot(neg_down, func(a,b,c,neg_down), 'b', label = 'Убывает меньше 0')
plt.legend()
plt.grid()
```
