Задача 1. Случайная непрерывная величина A имеет равномерное распределение на
промежутке (200, 800].
Найдите ее среднее значение и дисперсию.

математическое ожидание  находим по формуле  (a+b)/2, дисперсия = (b-a)**2/12

```python
M_t1 = (200 + 800)/2
D_t1 = ((800 - 200)**2)/12
print(f'среднее значение: {M_t1}, дисперсия: {D_t1}')
```

```python
b = ((12 * 0.2) ** 0.5) + 0.5
print(f'правая граница = {b:.2f}')
print(f'среднее взначение = {(0.5 + b) / 2}')
```

Задача 4. Рост взрослого населения города X имеет нормальное распределение, причем, средний рост равен 174 см, а среднее квадратическое отклонение равно 8 см. посчитайте, какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
1. больше 182 см?
2. больше 190 см?
3. от 166 см до 190 см?
4. от 166 см до 182 см?
5. от 158 см до 190 см?
6. не выше 150 см или не ниже 190 см?
7. не выше 150 см или не ниже 198 см?
8. ниже 166 см?

Задачу можно решить двумя способами: без использования сторонних библиотек (numpy, scipy, pandas и пр.), а затем проверить себя с помощью встроенных функций

Z= (x-mu) / si где  mu = 174, а si = 8

```python
mu = 174
si = 8
z1 = (182 - mu)/ si
print(f'z1 ={z1}')
z2 = (190 - mu)/ si
print(f'z2 ={z2}')
z31 = (166 - mu)/ si
z32 = (190 - mu)/ si
print(f'z3/1 ={z31}, z3/2 ={z32}')
z41 = (166 - mu)/ si
z42 = (182 - mu)/ si
print(f'z4/1 ={z41}, z4/2 ={z42}')
z51 = (158 - mu)/ si
z52 = (190 - mu)/ si
print(f'z5/1 ={z51}, z5/2 ={z52}')
z61 = (150 - mu)/ si
z62 = (190 - mu)/ si
print(f'z6/1 ={z61}, z5/2 ={z62}')
z71 = (150 - mu)/ si
z72 = (198 - mu)/ si
print(f'z7/1 ={z71}, z7/2 ={z72}')
z8 = (166 - mu)/ si
print(f'z8 ={z8}')
```

По таблице смотрим соответсвия:
1. больше 182 см? при z = 1 вероятность = 1 - 0.84134
2. больше 190 см? при z = 2 вероятность = 1 - 0.97725
3. от 166 см до 190 см? при z = -1 и z = 2 вероятность = 0.15866 и 0.97725
4. от 166 см до 182 см? при z = -1 и z = 1 вероятность = 0.15866 и 0.84134
5. от 158 см до 190 см? при z = -2 и z = 2 вероятность = 0.02275 и 0.97725
6. не выше 150 см или не ниже 190 см? при z = -3 и z = 2 вероятность = 0.00135 и 1 - 0.97725
7. не выше 150 см или не ниже 198 см? при z = -3 и z = 3 вероятность = 0.00135 и 1 - 0.99865
8. ниже 166 см? при z = -1 вероятность = 0.15866

```python
print(f'больше 182 см вероятность: {1 - 0.84134:.4f}')
print(f'больше 190 см вероятность: {1 - 0.97725:.4f}')
print(f'от 166 см до 190 см вероятность: {0.97725 - 0.15866:.4f}')
print(f'от 166 см до 182 см: {0.84134 - 0.15866:.4f}')
print(f'от 158 см до 190 см: {0.97725 - 0.02275:.4f}')
print(f'не выше 150 см или не ниже 190 см вероятность: {0.00135 + 1 - 0.97725:.4f}')
print(f'не выше 150 см или не ниже 198 см вероятность: {0.00135 + 1 - 0.99865:.4f}')
print(f'ниже 166 см вероятность: {0.15866:.4f}')
```

```python
print(f'больше 182 см вероятность: {1 - stats.norm.cdf(182, loc = 174, scale = 8):.4f}')
print(f'больше 190 см вероятность: {1 - stats.norm.cdf(190, loc = 174, scale = 8):.4f}')
print(f'от 166 см до 190 см вероятность: {stats.norm.cdf(190, loc = 174, scale = 8) - stats.norm.cdf(166, loc = 174, scale = 8):.4f}')
print(f'от 166 см до 182 см: {stats.norm.cdf(182, loc = 174, scale = 8) - stats.norm.cdf(166, loc = 174, scale = 8):.4f}')
print(f'от 158 см до 190 см: {stats.norm.cdf(190, loc = 174, scale = 8) - stats.norm.cdf(158, loc = 174, scale = 8):.4f}')
print(f'не выше 150 см или не ниже 190 см вероятность: {stats.norm.cdf(150, loc = 174, scale = 8) + 1 - stats.norm.cdf(190, loc = 174, scale = 8):.4f}')
print(f'не выше 150 см или не ниже 198 см вероятность: {stats.norm.cdf(150, loc = 174, scale = 8) + 1 - stats.norm.cdf(198, loc = 174, scale = 8):.4f}')
print(f'ниже 166 см вероятность: {stats.norm.cdf(166, loc = 174, scale = 8):.4f}')
```
