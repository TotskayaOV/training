```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('fifa_s2.csv')
df.info()
```

```python
df.isna().sum()/len(df)*100
```

сначала проводим поверхностную оценку всего объема данных. Затем проверяем значения в столбце Club

```python
df['Club'].unique()
```

при просмотре уникальных значений данных видно, что в данные попала нумераия из списков. Удаляем "1. " из данного столбца и приводим префикс клуба к единому формату.

```python
df['Club'] = df['Club'].str.replace('1. ', '')
df['Club'] = df['Club'].str.replace('Fc ', 'FC ')
```

```python
df['ID'].nunique()
```

проверяем строки с данными в котором дублируются значения по столбцу ID. Удостоверившись, что данные полностью совпадают, удаляем дубликаты

```python
duplicates = df[df.duplicated(subset='ID', keep=False)]
duplicates
```

```python
df = df.drop_duplicates(subset='ID')
df.info()
```

мы можем удалить дубликаты в которых совпадают значения по столбцам имя, клуб и дата присоединения к клубу (последнее, чтобы исключить варианты когда переходили из клуба в клуб)

```python
df = df.drop_duplicates(subset = ['Name', 'Club', 'Joined'])
rows_without_nationality = df[df['Nationality'].isnull()]
rows_without_nationality
```

Мы можем предположить что для каждого значения стобца Nationality соответствует такое же или мода значения для значений по столбцу Club. Также присвоим медианные значения столбцов пустым значениям в столбцах дохода и репутации.

```python
club_nationality_mapping = df[['Nationality', 'Club']].groupby('Club').agg(lambda x: x.value_counts().idxmax())
df['Nationality'] = df.apply(lambda row: club_nationality_mapping.loc[row['Club']][0] if pd.isna(row['Nationality']) else row['Nationality'], axis=1)
wage_median = df['Wage'].median()
df['Wage'] = df['Wage'].fillna(wage_median)
reputation_median = df['International Reputation']
df['International Reputation'] = df['International Reputation'].fillna(reputation_median)
df.info()
```

Изменяем формат столбца с датой окончания контракта. Создаем новый столбец с категориями возраста.

```python
df['Contract Valid Until'] = pd.to_datetime(df['Contract Valid Until'], format='%d/%m/%Y')

def splitting_age(age):
    if age < 20:
        return '< 20'
    elif 20 <= age < 30:
        return '20-29'
    elif 30 <= age <= 36:
        return '30-36'
    else:
        return '36 <'
    
df['period_age'] = df['Age'].apply(splitting_age)
df
```

удаляем строки с пустыми значениями

```python
df = df.dropna(subset=['Position', 'International Reputation', 'Release Clause', 'Value', 'Contract Valid Until'])
df.info()
```
